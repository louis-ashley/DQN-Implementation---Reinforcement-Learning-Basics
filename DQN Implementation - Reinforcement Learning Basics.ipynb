# This Third chapter focus on Deep Q-Network implementation

The implementation of the DQN is presented in this Notebook.

Next Figure shows the flowchart of the DQN, the different functions used in the flowchart are explained in this section.

A number of episodes to play (Nepisode) is selected; the training phase only begins when the
number of iterations performed (episode) exceeds a predefined limit number (Nstore).


Done is a value that means if the game is over or no.

The first part of the implementation is the creation and initialization of different elements
necessary for the operation of the program, 5 matrices that will serve as memory are created
and initialized with null values. 

These 5 matrices are respectively for initial states, actions,
rewards, next states and a binary to represent if the episode is completed (done). For rewards,
action and done matrices, the matrix is simply composed of a row and Nstore columns
corresponding to Nstore different situations that can be stored in the memory. As for states,
one value per situation is not enough, it is necessary to be able to place an entire image, so a
number of lines equivalent to the number of pixels in the image that will represent each state
is needed.

It is also necessary to create and initialize the two neural networks that will be used for the
operation of the DQN, these two NN’s must have the same structure (as many layers with as
many neurons per layer). The structure of the neural network for the final study of the project
is composed of 126 neurons for the input layer, followed by two layers of 60 neurons and
finally 5 output neurons for the 5 possible actions.



A first loop is created to repeat the DQN on the number of episodes (number of games in the
case of a game) desired.
A second loop is set up within the first, this loop continues until the end of an episode.
The next part consists of an action selection function as outlined in next figure, similar to Q-
Learning, a random value is taken and if it is higher than the Epsilon parameter the best action
will be chosen thanks to the NN and otherwise it will be a random action among the different
possible actions.

The next step takes place in the simulation of the environment; as seen in figure 23, the
current state as well as the action taken is used as input to this environment which will return
the reward, the next state and a binary to indicate whether the episode has ended or not.

The difference with Q-Learning is that this time no Q-Table is used, the neural network is
used to find the best action to take. A series of functions is used to transform the initial image
into a series of values compatible with a neural network.

At that moment a set of values is known: the current state, the action taken, the reward, the
next state and if the episode is over. 

As depicted in next figure, these different values are placed
in the different memory matrices initialized at the beginning. A function is added so that once
these matrices are full (Nstore iterations) the new values take the place of the old ones and
therefore the program rewrites to the old memory.
This, limits the size of the memory arrays and accelerates the speed of the algorithm.

Until then, these steps constitute the first part of the training that builds the memory seen in
theory.


Training is only applied if the memory has been implemented more than Nstore times.
For this training it is necessary to randomly select a certain number of memory layers, in our
case 32 layers are selected, these 32 layers will be used to modify the NN thanks to different
functions that use the two neural networks created during the initialization, which will not be
detailed. These successive modifications will make the NN tend to learn how to act to
optimize the score in the long term.


As represented in the following figure, to start this training, a &quot;batch&quot; is selected with 32 memory layers.
The layers are used one by one to update a second neural network. The network is therefore
updated 32 times with each iteration.

This is the last operation in the second loop.

Once all the episodes have been completed, the
program stops and provides the trained neural network for the application made.
